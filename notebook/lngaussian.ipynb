```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(42)
```

```python
Ntrials = 1024
means = 2. * np.ones(Ntrials)
tiny = 0.00003
variances = tiny * (0.5 + np.arange(Ntrials))
stds = np.sqrt(variances)
```

```python
Ndata = 2048
data = means[:,None] + stds[:,None] * np.random.normal(size=(Ntrials, Ndata))
empiricalmeans = np.mean(data, axis=1)
empiricallogmeans = np.mean(np.log(data), axis=1)
```

```python
plt.clf()
plt.plot(variances, empiricalmeans, "ko", alpha=0.25)
plt.plot(variances, means, "r-")
plt.xlabel("variance")
plt.ylabel("empirical mean of y")
```

```python
plt.clf()
plt.plot(variances, empiricallogmeans, "ko", alpha=0.25)
biases = -1. * variances / (2. * means * means)
plt.plot(variances, np.log(means) + biases, "r-")
plt.xlabel("variance")
plt.ylabel("empirical mean of ln(y)")
```

```python
correctedempiricallogmeans = empiricallogmeans - biases
plt.clf()
plt.plot(variances, correctedempiricallogmeans, "ko", alpha=0.25)
plt.plot(variances, np.log(means), "r-")
plt.xlabel("variance")
plt.ylabel("corrected empirical mean of ln(y)")
```
